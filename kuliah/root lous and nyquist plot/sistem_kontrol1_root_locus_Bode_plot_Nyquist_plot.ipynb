```python
#Nomor 1B Nyquist Plot
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

num1 = [1, 5, 6]
den1 = [1, 0,-1]

num2 = [1]
den2 = [1,15,50,0]

num3 = [1]
den3 = [1,6,11,6]


W1 = ctrl.tf(num1,den1)
print(W1)

W2 = ctrl.tf(num2,den2)
print(W2)

W3 = ctrl.tf(num3,den3)
print(W3)

count1, contour1 = ctrl.nyquist_plot(W1, return_contour=True, arrows = 2,
                                   arrow_size=12, start_marker='o', start_marker_size=12, linewidth=4, plot=True)

plt.show()
```

```python
#Nomor 1B kestabilan unity feedback
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# Define transfer functions
num1 = [1, 5, 6]
den1 = [1, 0, -1]
W1 = ctrl.TransferFunction(num1, den1)

num2 = [1]
den2 = [1, 15, 50, 0]
W2 = ctrl.TransferFunction(num2, den2)

num3 = [1]
den3 = [1, 6, 11, 6]
W3 = ctrl.TransferFunction(num3, den3)

# Print transfer functions
print("W1:", W1)
print("W2:", W2)
print("W3:", W3)

# Calculate and print poless and zeross for open loop systems
print("Poless of W1:", W1.poles())
print("Zeross of W1:", W1.zeros())

print("Poless of W2:", W2.poles())
print("Zeross of W2:", W2.zeros())

print("Poless of W3:", W3.poles())
print("Zeross of W3:", W3.zeros())

# Closed loop transfer functions with unity feedback
W1_closed_loop = ctrl.feedback(W1, 1)
W2_closed_loop = ctrl.feedback(W2, 1)
W3_closed_loop = ctrl.feedback(W3, 1)

# Print closed loop transfer functions
print("Closed loop W1:", W1_closed_loop)
print("Closed loop W2:", W2_closed_loop)
print("Closed loop W3:", W3_closed_loop)

# Calculate and print poless of closed loop systems (to determine stability)
print("Poless of closed loop W1:", W1_closed_loop.poles())
print("Poless of closed loop W2:", W2_closed_loop.poles())
print("Poless of closed loop W3:", W3_closed_loop.poles())

# Plot step response to observe stability in time domain
time = np.linspace(0, 10, 1000)

t1, y1 = ctrl.step_response(W1_closed_loop, time)
t2, y2 = ctrl.step_response(W2_closed_loop, time)
t3, y3 = ctrl.step_response(W3_closed_loop, time)

plt.figure()

plt.subplot(3, 1, 1)
plt.plot(t1, y1)
plt.title('Step Response of Closed Loop W1')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t2, y2)
plt.title('Step Response of Closed Loop W2')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t3, y3)
plt.title('Step Response of Closed Loop W3')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
